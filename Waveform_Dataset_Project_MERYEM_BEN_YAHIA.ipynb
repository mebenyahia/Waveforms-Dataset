###### **1. Tuning the best K of a kNN classifier by Cross-validation**

 In this experiment, the objective was to plot the accuracies over the validation subset with respect to k, from 4000 randomly drawn training examples (apart 1000 waves for the test set, to find the best k).

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X = waveformdata.iloc[:, :-1]  # features
y = waveformdata.iloc[:, -1]   # target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000, train_size=3999, random_state=42)

k_values = range(1, 64) #using the square root rule: sqrt(3999) is approximately 64 rounded-up

# cross-validation scores for each k
cv_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    cv_scores.append(scores.mean())

# accuracies over the validation subset with respect to k
plt.figure(figsize=(12, 8))
plt.scatter(k_values, cv_scores, color='b', label='Cross-Validated Accuracy')

# Optimal Bayes classification rate (86%)
optimal_bayes_accuracy = 0.86
plt.axhline(y=optimal_bayes_accuracy, color='g', linestyle='--', label='Optimal Bayes Accuracy (86%)')

plt.title('Accuracy vs. K for kNN', fontsize=16)
plt.xlabel('Number of Neighbors (K)', fontsize=14)
plt.ylabel('Cross-Validated Accuracy', fontsize=14)
plt.xticks(np.arange(min(k_values), max(k_values)+1, 5.0))
plt.legend()
plt.grid(True)
plt.show()
```

###### **2. Reducing the complexity of the kNN classifier**

Algo from the slides: Using the first algorithm studied in class, the preliminary step is to remove from S (waveform dataset) the outliers and the examples of the bayes error region.

```python
def iterative_cleaning(S, random_state=42):
    # Split randomly S into two subsets S1 and S2
    S1, S2 = train_test_split(S, test_size=0.5, random_state=random_state)

    previous_len_S1 = 0
    previous_len_S2 = 0

    # While no stabilization of S1 and S2
    while len(S1) != previous_len_S1 or len(S2) != previous_len_S2:
        previous_len_S1, previous_len_S2 = len(S1), len(S2)

        # Classify S1 with S2 using the 1-NN rule
        knn = KNeighborsClassifier(n_neighbors=1)
        knn.fit(S2.iloc[:, :-1], S2.iloc[:, -1])
        S1_predictions = knn.predict(S1.iloc[:, :-1])

        # Remove from S1 the misclassified instances
        S1 = S1[S1_predictions == S1.iloc[:, -1].values]

        # Classify S2 with the new set S1 using the 1-NN rule
        knn.fit(S1.iloc[:, :-1], S1.iloc[:, -1])
        S2_predictions = knn.predict(S2.iloc[:, :-1])

        # Remove from S2 the misclassified instances
        S2 = S2[S2_predictions == S2.iloc[:, -1].values]

    # Combine S1 and S2
    Scleaned = pd.concat([S1, S2])
    return Scleaned


cleaned_data = iterative_cleaning(waveformdata)
cleaned_data.reset_index(drop=True, inplace=True)
cleaned_data.info()
```

Applying the second step: removing irrelevant examples from the cleaned data.

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

def remove_irrelevant_examples(S, random_state=42):
    STORAGE = pd.DataFrame()
    DUSTBIN = pd.DataFrame()

    # Start with one random example in STORAGE
    STORAGE = STORAGE.append(S.sample(n=1, random_state=random_state))
    S = S.drop(STORAGE.index)

    # Iterate until STORAGE stabilizes
    while True:
        previous_STORAGE = STORAGE.copy()
        for index, example in S.iterrows():

            # Train 1-NN classifier with current STORAGE
            knn = KNeighborsClassifier(n_neighbors=1)
            knn.fit(STORAGE.iloc[:, :-1], STORAGE.iloc[:, -1])

            # Classify current example
            prediction = knn.predict([example[:-1]])
            if prediction == example[-1]:
                DUSTBIN = DUSTBIN.append(example)
            else:
                STORAGE = STORAGE.append(example)

        # STORAGE didn't change
        if STORAGE.equals(previous_STORAGE):
            break

        DUSTBIN = pd.DataFrame()  # reset DUSTBIN

    return STORAGE.reset_index(drop=True)

final_reduced_data = remove_irrelevant_examples(cleaned_data)
```

###### **3. Speeding up the calculation of the 1NN (with a brute force 1NN)**

```python
class KDTreeNode:
    def __init__(self, point, label, left=None, right=None):
        self.point = point
        self.label = label
        self.left = left
        self.right = right

def build_kdtree(points, labels, depth=0):
    n = len(points)
    if n == 0:
        return None

    axis = depth % len(points[0])
    sorted_points = sorted(zip(points, labels), key=lambda x: x[0][axis])
    median = n // 2

    return KDTreeNode(
        point=sorted_points[median][0],
        label=sorted_points[median][1],
        left=build_kdtree([x[0] for x in sorted_points[:median]], [x[1] for x in sorted_points[:median]], depth + 1),
        right=build_kdtree([x[0] for x in sorted_points[median + 1:]], [x[1] for x in sorted_points[median + 1:]], depth + 1)
    )

def find_nearest_neighbor(root, point, depth=0, best=None):
    if root is None:
        return best

    axis = depth % len(point)
    next_best = None
    next_branch = None

    if best is None or np.linalg.norm(np.array(point) - np.array(root.point)) < np.linalg.norm(np.array(point) - np.array(best.point)):
        next_best = root
    else:
        next_best = best

    if point[axis] < root.point[axis]:
        next_branch = find_nearest_neighbor(root.left, point, depth + 1, next_best)
    else:
        next_branch = find_nearest_neighbor(root.right, point, depth + 1, next_best)

    if next_branch is None:
        return next_best

    if np.linalg.norm(np.array(point) - np.array(next_branch.point)) < np.linalg.norm(np.array(point) - np.array(next_best.point)):
        return next_branch
    else:
        return next_best


data_waveformdata = waveformdata.iloc[:, :-1].values
labels_waveformdata = waveformdata.iloc[:, -1].values


kd_tree_waveformdata = build_kdtree(data_waveformdata, labels_waveformdata)

test_point_waveformdata = data_waveformdata[0]
nearest_neighbor_waveformdata = find_nearest_neighbor(kd_tree_waveformdata, test_point_waveformdata)
```

```python
import time

start_time_build = time.time()
kd_tree_waveformdata = build_kdtree(data_waveformdata, labels_waveformdata)

duration_build = time.time() - start_time_build
start_time_search = time.time()

nearest_neighbor_waveformdata = find_nearest_neighbor(kd_tree_waveformdata, test_point_waveformdata)
duration_search = time.time() - start_time_search

build_time_result = f"KD-Tree Building Time: {duration_build:.4f} seconds"
search_time_result = f"1-Nearest Neighbor Search Time: {duration_search:.4f} seconds"
nearest_neighbor_result = f"Nearest Neighbor Point: {nearest_neighbor_waveformdata.point}, Label: {nearest_neighbor_waveformdata.label}"

build_time_result, search_time_result, nearest_neighbor_result
```

```python
def brute_force_1nn(data, labels, query_point):
    min_distance = float('inf')
    nearest_neighbor = None
    nearest_label = None

    for point, label in zip(data, labels):
        distance = np.linalg.norm(np.array(query_point) - np.array(point))
        if distance < min_distance:
            min_distance = distance
            nearest_neighbor = point
            nearest_label = label

    return nearest_neighbor, nearest_label

start_time_brute_force = time.time()
nearest_neighbor_brute_force, nearest_label_brute_force = brute_force_1nn(data_waveformdata, labels_waveformdata, test_point_waveformdata)
duration_brute_force = time.time() - start_time_brute_force

brute_force_time_result = f"Brute-force 1NN Search Time: {duration_brute_force:.4f} seconds"
brute_force_time_result
```

```python
def triangle_inequality_1nn(data, labels, query_point):
    best_distance = float('inf')
    nearest_neighbor = None
    nearest_label = None

    for point, label in zip(data, labels):
        if abs(np.linalg.norm(np.array(query_point)) - np.linalg.norm(np.array(point))) >= best_distance:
            continue

        distance = np.linalg.norm(np.array(query_point) - np.array(point))
        if distance < best_distance:
            best_distance = distance
            nearest_neighbor = point
            nearest_label = label

    return nearest_neighbor, nearest_label

start_time_triangle_inequality = time.time()
nearest_neighbor_triangle_inequality, nearest_label_triangle_inequality = triangle_inequality_1nn(data_waveformdata, labels_waveformdata, test_point_waveformdata)
duration_triangle_inequality = time.time() - start_time_triangle_inequality

triangle_inequality_time_result = f"Triangle Inequality Optimized 1NN Search Time: {duration_triangle_inequality:.4f} seconds"
triangle_inequality_time_result
```

###### **4. Generating artificial imbalancy, and turning K using the F-measure**

```python
X = waveformdata.iloc[:, :-1]
y = waveformdata.iloc[:, -1]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=1000, random_state=42)

class_0 = X_train[y_train == 0]
class_1 = X_train[y_train == 1].sample(frac=0.5, random_state=42)  # -50%
class_2 = X_train[y_train == 2].sample(frac=0.25, random_state=42) # -75%

X_train_imbalanced = pd.concat([class_0, class_1, class_2])
y_train_imbalanced = pd.concat([y_train[y_train == 0], y_train[y_train == 1].loc[class_1.index], y_train[y_train == 2].loc[class_2.index]])


knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_imbalanced, y_train_imbalanced)

y_pred = knn.predict(X_test)

accuracy_imbalanced = accuracy_score(y_test, y_pred)
accuracy_imbalanced
```

```python
knn_balanced = KNeighborsClassifier(n_neighbors=1)
knn_balanced.fit(X_train, y_train)
y_pred_balanced = knn_balanced.predict(X_test)

accuracy_balanced = accuracy_score(y_test, y_pred_balanced)
accuracy_balanced
```

```python
from sklearn.metrics import f1_score

k_values = range(1, 64)

f1_scores = []
accuracies = []

for k in k_values:
    knn_k = KNeighborsClassifier(n_neighbors=k)
    knn_k.fit(X_train, y_train)
    y_pred_k = knn_k.predict(X_test)

    f1 = f1_score(y_test, y_pred_k, average='weighted')
    accuracy = accuracy_score(y_test, y_pred_k)

    f1_scores.append(f1)
    accuracies.append(accuracy)

best_k_f1 = k_values[np.argmax(f1_scores)]
best_f1_score = max(f1_scores)
best_accuracy_for_best_f1 = accuracies[np.argmax(f1_scores)]

best_k_f1, best_f1_score, best_accuracy_for_best_f1
```

```python
from imblearn.over_sampling import SMOTE
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

f_measure_list = []
accuracy_list = []

for k in k_values:
    smote_balancer = SMOTE(sampling_strategy='minority')
    balanced_X_train, balanced_y_train = smote_balancer.fit_resample(X_train, y_train)

    classifier = KNeighborsClassifier(n_neighbors=k)
    classifier.fit(balanced_X_train, balanced_y_train)

    predictions = classifier.predict(X_test)
    f_measure_value = f1_score(y_test, predictions, average='weighted')
    test_accuracy = classifier.score(X_test, y_test)

    f_measure_list.append(f_measure_value)
    accuracy_list.append(test_accuracy)
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(k_values, f_measure_list, color='red', label='F-measure')
plt.scatter(k_values, accuracy_list, color='blue', label='Accuracy')
plt.axhline(y=optimal_bayes_accuracy, color='g', linestyle='--', label='Optimal Bayes Accuracy (86%)')
plt.xlabel('Value of k (Number of Neighbors)', fontsize=12)
plt.ylabel('Performance Metrics', fontsize=12)
plt.title("Performance Comparison: Accuracy vs F-measure for k", fontsize=14)
plt.grid(True)
plt.legend()
plt.show()
```
